```python
import os
import pickle
import PIL.Image
import numpy as np
import dnnlib
import dnnlib.tflib as tflib
import config
from encoder.generator_model import Generator
import matplotlib.pyplot as plt
```

Calling the generator of pictures

```python
#URL_FFHQ = 'https://drive.google.com/uc?id=1MEGjdvVpUsu1jB4zrXZN7Y4kBBOzizDQ'
#URL_FFHQ = 'https://drive.google.com/uc?id=1MEGjdvVpUsu1jB4zrXZN7Y4kBBOzizDQ'
URL_FFHQ = 'https://drive.google.com/uc?id=1PrjiBIXQ-tEa3l3vAtBR7FLi-n9WnVB8'
tflib.init_tf()
with dnnlib.util.open_url(URL_FFHQ, cache_dir=config.cache_dir) as f:
    generator_network, discriminator_network, Gs_network = pickle.load(f)
```

```python
generator = Generator(Gs_network, batch_size=1, randomize_noise=False)
```

```python
def generate_image(latent_vector):
    latent_vector = latent_vector.reshape((1, 18, 512))
    generator.set_dlatents(latent_vector)
    img_array = generator.generate_images()[0]
    img = PIL.Image.fromarray(img_array, 'RGB')
    return img.resize((256, 256))

def move_and_show(latent_vector, direction, coeffs):
    fig,ax = plt.subplots(1, len(coeffs), figsize=(15, 10), dpi=80)
    for i, coeff in enumerate(coeffs):
        new_latent_vector = latent_vector.copy()
        new_latent_vector[:8] = (latent_vector + coeff*direction)[:8]
        ax[i].imshow(generate_image(new_latent_vector))
        ax[i].set_title('Coeff: %0.1f' % coeff)
    [x.axis('off') for x in ax]
    plt.show()

def save_modified_image(latent_vector, direction, coeff, output):
  new_latent_vector = latent_vector.copy()
  new_latent_vector[:8] = (latent_vector + coeff*direction)[:8]
  img = generate_image(new_latent_vector)
  img = img.save(output) 
```

```python
#cover:  small picture (just the face)
#template: big picture (al the frame)
def transform(template, cover, destination_points, face_dimension):
  with Image(filename=cover) as cover, Image(filename=template) as template:
      w, h = cover.size
      print(f"Size of cover: {(w, h)}")
      print(f"Size of template: {(template.size)}")
      cover.virtual_pixel = 'transparent'
      source_points = (
          (0, 0),
          (0, h),
          (w, h),
          (w, 0)
      )
      # destination_points = (
      #     (628+78.37203406,  35.24937345),
      #     (628+577.65062655,  62.72203406),
      #     (628+550.17796594, 562.00062655),
      #     (628+50.89937345, 534.52796594)
      # )
      order = chain.from_iterable(zip(source_points, destination_points))
      print(f"order: {order}")
      arguments = list(chain.from_iterable(order))
      print(f"arguments: {arguments}")
      cover.save(filename='./cover_before.jpg')
      cover.distort('perspective', arguments)
      # cover.ALPHA_CHANNEL_TYPES = 'activate'
      x_center = int((destination_points[0][0]+ destination_points[2][0])/2)
      y_center = int((destination_points[0][1]+ destination_points[2][1])/2)
      face_w, face_h = face_dimension
      center = (x_center, y_center)
      left = x_center - (face_w // 2)
      top = y_center - (face_h // 2)
      cover.crop(left= left, top= top, width=face_w, height=face_h)
      cover.save(filename='./cover_after.jpg')
      print(center)
      # Overlay cover onto template and save
      # template.composite_channel(channel= 'magenta', 
      #                            image  = cover,
      #                            operator = 'blend',
      #                            left=0,
      #                            top=0)
      template.composite(cover,left=0,top=0)
      template.save(filename='./video/result.png')
```
